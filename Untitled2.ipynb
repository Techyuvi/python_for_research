```python
class NewList(list):
   def remove_max(self):
     self.remove(max(self))
   def append_sum(self):
     self.append(sum(self))

x = NewList([1,2,3])
print(x)
print(type(x))
while max(x) < 10:
   x.remove_max()
   x.append_sum()
print(type(x))
```
