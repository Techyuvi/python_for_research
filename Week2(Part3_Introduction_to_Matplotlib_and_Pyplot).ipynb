```python
y1=x*2.0
y2=x*1.5
plot.plot(x,y1,"gs-",linewidth=1,markersize=4,label="first data")
plot.plot(x,y2,"rs-",linewidth=1,markersize=4,label="second data")
plot.xlabel("X")
plot.ylabel("Y")
plot.axis([0,10,0,20])
plot.legend(loc="upper left")
plot.savefig("Myfigure.pdf")
print(y2)
print(y1)
```

```python
y1=x*2.0
y2=x*1.5
plot.loglog(x,y1,"gs-",linewidth=1,markersize=4,label="first data")
plot.loglog(x,y2,"rs-",linewidth=1,markersize=4,label="second data")
plot.xlabel("X")
plot.ylabel("Y")
plot.axis([0,10,0,20])
plot.legend(loc="upper left")
plot.savefig("Mylogfigurel.pdf")
print(y2)
print(y1)
```
