```python
#dense graph
N=20
p=0.2

Em = nx.Graph()
Em.add_nodes_from(range(N))
for n1 in Em.nodes():
    for n2 in Em.nodes():
        if bernoulli.rvs(p=p):
            Em.add_edge(n1,n2)
            
    
```

```python
nx.draw(Em,with_labels=True,edge_color="red",node_color="green")
```

```python
Em.number_of_nodes()
```

```python
#less dense graph

N=20
p=0.2
def Em_graph(N,p):
    Em = nx.Graph()
    Em.add_nodes_from(range(N))
    for n1 in Em.nodes():
        for n2 in Em.nodes():
            if n1<n2 and bernoulli.rvs(p=p):
                Em.add_edge(n1,n2)
    return Em

nx.draw(Em_graph(N,p),with_labels=True,edge_color="red",node_color="green");
Em.degree()
plt.savefig("Erdos-Renyi_Graph.pdf");
```

```python
####### ploting Degree distribution 
G = Em_graph(50,0.08)
def plot_degree_distrubtion(G):
    plt.hist(list(dict(G.degree()).values()), histtype = "step")
    plt.xlabel("Degree $k$")
    plt.ylabel("$Pk$")
    plt.title("degree distrubtion")
```

```python
plot_degree_distrubtion(G)
plt.savefig("degree distrubtion.pdf");
```
